```python
dicc = {}


# Porcentajes de mejora tabu/greedy
d50 = {}
d75 = {}
d100 = {}

# Tiempos para cada numero de clientes
d50time = {}
d75time = {}
d100time = {}

# Nombres de filas para cada numero de clientes
filas50 = []
filas75 = []
filas100 = []

column = "({0}, {1})"
for instancia in instances:
    dicc[instancia] = []
    for ite in iterations:
        for le in lenTabu:
            arch = open(directory.format(instancia, ite, le) + ".out", 'r')
            
            numVisitedClients = arch.readline().strip()
            numVehicleUsed = arch.readline().strip()
            
            clients = arch.readline().strip()
            AFS = arch.readline().strip()
            greedyQuality = arch.readline().strip()
            tabuQuality = arch.readline().strip()
            execTime = arch.readline().strip()
            dicc[instancia].append((clients, AFS, numVisitedClients, numVehicleUsed, greedyQuality, tabuQuality, execTime))
            
            k = ab.format(instancia)
            if(clients == "50"):
                if(k not in filas50):
                    filas50.append(k)
                if(column.format(ite,le) not in d50):
                    d50[column.format(ite, le)] = []
                d50[column.format(ite,le)].append(format(round((1-(float(tabuQuality)/float(greedyQuality)))*100,2),'.2f'))
                
                if(column.format(ite,le) not in d50time):
                    d50time[column.format(ite, le)] = []
                d50time[column.format(ite,le)].append(float(execTime))
            
            elif(clients == "75"):
                if(k not in filas75):
                    filas75.append(k)
                if(column.format(ite,le) not in d75):
                    d75[column.format(ite, le)] = []
                d75[column.format(ite,le)].append(format(round((1-(float(tabuQuality)/float(greedyQuality)))*100,2),'.2f'))
                
                if(column.format(ite,le) not in d75time):
                    d75time[column.format(ite, le)] = []
                d75time[column.format(ite,le)].append(float(execTime))
                
            elif(clients == "100"):
                if(k not in filas100):
                    filas100.append(k)
                if(column.format(ite,le) not in d100):
                    d100[column.format(ite, le)] = []
                d100[column.format(ite,le)].append(format(round((1-(float(tabuQuality)/float(greedyQuality)))*100,2),'.2f'))
                
                if(column.format(ite,le) not in d100time):
                    d100time[column.format(ite, le)] = []
                d100time[column.format(ite,le)].append(float(execTime))
            
            arch.close()
            
data_clientes50 = []
data_clientes75 = []
data_clientes100 = []
data_fuel50 =  []
data_fuel75 =  []
data_fuel100 =  []

for i in dicc:
    clients = dicc[i][0][0]
    fuel = dicc[i][0][1]
    if(clients == "50"):
        data_clientes50.append(clients)
        data_fuel50.append(fuel)
    elif(clients == "75"):
        data_clientes75.append(clients)
        data_fuel75.append(fuel)
    elif(clients == "100"):
        data_clientes100.append(clients)
        data_fuel100.append(fuel)

dicc
```

```python
df50 = DataFrame(data=d50, index=filas50)  
df50.insert(0,"#clients", data_clientes50)
df50.insert(1,"#afs", data_fuel50)

df50.style.highlight_max(subset = df50.columns[-12:], color = 'lightgreen', axis = 1)
```

```python
# En promedio los mejores resultados se obtuvieron para la configuracion 
# numIte = 55, lenList = 13
# Se tabulan los datos para las instancias  AB101,AB102...AB204
bestIndex1 = 10 # mejor indice de soluciones

best50 = {}
best50["#clientes"] = []
best50["#afs"] = []
best50["#clientes visitados"] = []
best50["#vehiculos enviados"] = []
best50["distancia Greedy"] = []
best50["distancia Tabu"] = []
best50["tiempo [s]"] = []
    
for inst in filas50:
    num = int(inst[2:])
    cl, stat, numC, numV, gQ, tQ, eTime = dicc[num][bestIndex1]
    best50["#clientes"].append(cl)
    best50["#afs"].append(stat)
    best50["#vehiculos enviados"].append(numV)
    best50["#clientes visitados"].append(numC)
    best50["distancia Greedy"].append(gQ)
    best50["distancia Tabu"].append(tQ)
    best50["tiempo [s]"].append(eTime)

dfBest50 = DataFrame(data=best50, index=filas50) 
dfBest50
```
